```python
import numpy as np
import matplotlib.pyplot as plt

NER_Chewie2016 = np.load('./data_NER/Fig4/NPZ_PMd_S1/NER_Chewie2016.npz')
Cebra_Chewie2016 = np.load('./data_NER/Fig4/NPZ_PMd_S1/Cebra_Chewie2016.npz')
piVAE_Chewie2016 = np.load('./data_NER/Fig4/NPZ_PMd_S1/piVAE_Chewie2016.npz')
UMAP_Chewie2016 = np.load('./data_NER/Fig4/NPZ_PMd_S1/UMAP_Chewie2016.npz')
UMAP_label_Chewie2016 = np.load('./data_NER/Fig4/NPZ_PMd_S1/UMAP_label_Chewie2016.npz')

NER_Han2017 = np.load('./data_NER/Fig4/NPZ_PMd_S1/NER_Han2017.npz')
Cebra_Han2017 = np.load('./data_NER/Fig4/NPZ_PMd_S1/Cebra_Han2017.npz')
piVAE_Han2017 = np.load('./data_NER/Fig4/NPZ_PMd_S1/piVAE_Han2017.npz')
UMAP_Han2017 = np.load('./data_NER/Fig4/NPZ_PMd_S1/UMAP_Han2017.npz')
UMAP_label_Han2017 = np.load('./data_NER/Fig4/NPZ_PMd_S1/UMAP_label_Han2017.npz')

labels=np.concatenate((NER_Chewie2016['date_subjects'], NER_Han2017['date_subjects']))
labels = np.unique(labels)
n_file = len(labels)
plot_xticks = np.arange(n_file)

NER_pos_C = NER_Chewie2016['pos_R_2D']
Cebra_pos_C = Cebra_Chewie2016['pos_R_2D']
piVAE_pos_C = piVAE_Chewie2016['pos_R_2D']
UMAP_label_pos_C = UMAP_label_Chewie2016['pos_R_2D']
UMAP_pos_C = UMAP_Chewie2016['pos_R_2D']

NER_vel_C = NER_Chewie2016['vel_R_2D']
Cebra_vel_C = Cebra_Chewie2016['vel_R_2D']
piVAE_vel_C = piVAE_Chewie2016['vel_R_2D']
UMAP_label_vel_C = UMAP_label_Chewie2016['vel_R_2D']
UMAP_vel_C = UMAP_Chewie2016['vel_R_2D']

NER_pos_H = NER_Han2017['pos_R_2D']
Cebra_pos_H = Cebra_Han2017['pos_R_2D']
piVAE_pos_H = piVAE_Han2017['pos_R_2D']
UMAP_label_pos_H = UMAP_label_Han2017['pos_R_2D']
UMAP_pos_H = UMAP_Han2017['pos_R_2D']

NER_vel_H = NER_Han2017['vel_R_2D']
Cebra_vel_H = Cebra_Han2017['vel_R_2D']
piVAE_vel_H = piVAE_Han2017['vel_R_2D']
UMAP_label_vel_H = UMAP_label_Han2017['vel_R_2D']
UMAP_vel_H = UMAP_Han2017['vel_R_2D']
```

```python
time_resolution = 30
dur = 40
time_bins = np.arange(0, dur * time_resolution, time_resolution)
plt.rcParams.update({'font.size': 15, 'font.family': 'DejaVu Sans'})
mean_trace = np.zeros((dur, 6))
fig = plt.figure(figsize=(7,5))
ax = plt.subplot(111)

mean_values = np.mean(NER_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(NER_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='NER '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 0] = mean_values

mean_values = np.mean(Cebra_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(Cebra_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='Cebra '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 1] = mean_values

mean_values = np.mean(piVAE_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(piVAE_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='piVAE '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 2] = mean_values

mean_values = np.mean(UMAP_label_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(UMAP_label_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='#9467bd',label='UMAP w/ label '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
# mean_trace[:, 3] = mean_values ### all "1" can't be used for correlation matrix

mean_values = np.mean(UMAP_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(UMAP_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='#d62728', label='UMAP w/o '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 3] = mean_values

Chewie20161014 = np.load('./data_NER/Fig4/NPZ_M1/Chewie_161014_embed_20000itr_M1.npz') ### M1+PMd have same XY
Chewie20161014_XYTarget = Chewie20161014['continuous_index_train']
Chewie20161014_XYTarget.shape
X_vel = np.abs(Chewie20161014_XYTarget[:, 0]).reshape((-1, dur))
mean_values = np.mean(X_vel, axis=0)
mean_values = mean_values/np.max(mean_values)*100
sem_values = np.std(X_vel, axis=0) / np.sqrt(X_vel.shape[0])
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='gray', label='X-velocities '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 4] = mean_values

Y_vel = np.abs(Chewie20161014_XYTarget[:, 1]).reshape((-1, dur))
mean_values = np.mean(Y_vel, axis=0)
mean_values = mean_values/np.max(mean_values)*100
sem_values = np.std(Y_vel, axis=0) / np.sqrt(Y_vel.shape[0])
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='black', label='Y-velocities '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 5] = mean_values
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Time from go cue (ms)')
plt.ylabel('Direction classification accuracy (%)')
# plt.title('Mean and SD across sessions: Chewie2016 PMd')
# ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.grid(False)
plt.savefig('./data_NER/Fig4/Chewie2016_Direction_accuracy_across_time_PMd.pdf', bbox_inches='tight')

##### as a reference only #####
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 7))
df_mean_trace = pd.DataFrame(mean_trace, columns=[f'Feature_{i+1}' for i in range(6)])
corr_matrix = df_mean_trace.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
labels_CC = ['NER', 'Cebra', 'piVAE', 'UMAP w/o label', 'X-vel', 'Y-vel']
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='gray', vmin=0, vmax=1,
           xticklabels=labels_CC, yticklabels=labels_CC)
# plt.title('Correlation Matrix (Upper Triangle without Diagonal)')
plt.show()
```
