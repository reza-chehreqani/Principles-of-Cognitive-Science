```python
import numpy as np
import matplotlib.pyplot as plt

NER_Chewie2016 = np.load('./data_NER/Fig4/NPZ_M1/NER_Chewie2016.npz')
Cebra_Chewie2016 = np.load('./data_NER/Fig4/NPZ_M1/Cebra_Chewie2016.npz')
piVAE_Chewie2016 = np.load('./data_NER/Fig4/NPZ_M1/piVAE_Chewie2016.npz')
UMAP_Chewie2016 = np.load('./data_NER/Fig4/NPZ_M1/UMAP_Chewie2016.npz')
UMAP_label_Chewie2016 = np.load('./data_NER/Fig4/NPZ_M1/UMAP_label_Chewie2016.npz')

NER_Chewie2015 = np.load('./data_NER/Fig4/NPZ_M1/NER_Chewie2015.npz')
Cebra_Chewie2015 = np.load('./data_NER/Fig4/NPZ_M1/Cebra_Chewie2015.npz')
piVAE_Chewie2015 = np.load('./data_NER/Fig4/NPZ_M1/piVAE_Chewie2015.npz')
UMAP_Chewie2015 = np.load('./data_NER/Fig4/NPZ_M1/UMAP_Chewie2015.npz')
UMAP_label_Chewie2015 = np.load('./data_NER/Fig4/NPZ_M1/UMAP_label_Chewie2015.npz')

labels=np.concatenate((NER_Chewie2016['date_subjects'],NER_Chewie2015['date_subjects']))
n_file = len(labels)
plot_xticks = np.arange(n_file)

NER_pos = np.vstack((NER_Chewie2016['pos_R_2D'], NER_Chewie2015['pos_R_2D']))
Cebra_pos = np.vstack((Cebra_Chewie2016['pos_R_2D'], Cebra_Chewie2015['pos_R_2D']))
piVAE_pos = np.vstack((piVAE_Chewie2016['pos_R_2D'], piVAE_Chewie2015['pos_R_2D']))
UMAP_pos = np.vstack((UMAP_Chewie2016['pos_R_2D'], UMAP_Chewie2015['pos_R_2D']))
UMAP_label_pos = np.vstack((UMAP_label_Chewie2016['pos_R_2D'], UMAP_label_Chewie2015['pos_R_2D']))

NER_vel = np.vstack((NER_Chewie2016['vel_R_2D'], NER_Chewie2015['vel_R_2D']))
Cebra_vel = np.vstack((Cebra_Chewie2016['vel_R_2D'], Cebra_Chewie2015['vel_R_2D']))
piVAE_vel = np.vstack((piVAE_Chewie2016['vel_R_2D'], piVAE_Chewie2015['vel_R_2D']))
UMAP_vel = np.vstack((UMAP_Chewie2016['vel_R_2D'], UMAP_Chewie2015['vel_R_2D']))
UMAP_label_vel = np.vstack((UMAP_label_Chewie2016['vel_R_2D'], UMAP_label_Chewie2015['vel_R_2D']))

NER_acc_peak = np.vstack((NER_Chewie2016['peak_acc_2D'], NER_Chewie2015['peak_acc_2D']))
Cebra_acc_peak = np.vstack((Cebra_Chewie2016['peak_acc_2D'], Cebra_Chewie2015['peak_acc_2D']))
piVAE_acc_peak = np.vstack((piVAE_Chewie2016['peak_acc_2D'], piVAE_Chewie2015['peak_acc_2D']))
UMAP_acc_peak = np.vstack((UMAP_Chewie2016['peak_acc_2D'], UMAP_Chewie2015['peak_acc_2D']))
UMAP_label_acc_peak = np.vstack((UMAP_label_Chewie2016['peak_acc_2D'], UMAP_label_Chewie2015['peak_acc_2D']))

NER_acc_avg = np.vstack((NER_Chewie2016['acc_time_2D'], NER_Chewie2015['acc_time_2D']))
Cebra_acc_avg = np.vstack((Cebra_Chewie2016['acc_time_2D'], Cebra_Chewie2015['acc_time_2D']))
piVAE_acc_avg = np.vstack((piVAE_Chewie2016['acc_time_2D'], piVAE_Chewie2015['acc_time_2D']))
UMAP_acc_avg = np.vstack((UMAP_Chewie2016['acc_time_2D'], UMAP_Chewie2015['acc_time_2D']))
UMAP_label_acc_avg = np.vstack((UMAP_label_Chewie2016['acc_time_2D'], UMAP_label_Chewie2015['acc_time_2D']))
```

```python
time_resolution = 30
dur = 40
time_bins = np.arange(0, dur * time_resolution, time_resolution)
plt.rcParams.update({'font.size': 12, 'font.family': 'DejaVu Sans'})
mean_trace = np.zeros((dur, 6))
fig = plt.figure(figsize=(7,5))
ax = plt.subplot(111)

mean_values = np.mean(NER_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(NER_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='NER '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 0] = mean_values

mean_values = np.mean(Cebra_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(Cebra_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='Cebra '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 1] = mean_values

mean_values = np.mean(piVAE_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(piVAE_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, label='piVAE '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 2] = mean_values

mean_values = np.mean(UMAP_label_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(UMAP_label_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='#9467bd',label='UMAP w/ label '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
# mean_trace[:, 3] = mean_values ### all "1" can't be used for correlation matrix

mean_values = np.mean(UMAP_Chewie2016['acc_time_2D'][:, 0, :], axis=0)
sem_values = np.std(UMAP_Chewie2016['acc_time_2D'][:, 0, :], axis=0) / 1 
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='#d62728', label='UMAP w/o '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 3] = mean_values

Chewie20161014 = np.load('./data_NER/Fig4/NPZ_M1/Chewie_161014_embed_20000itr_M1.npz')
Chewie20161014_XYTarget = Chewie20161014['continuous_index_train']
Chewie20161014_XYTarget.shape
X_vel = np.abs(Chewie20161014_XYTarget[:, 0]).reshape((-1, dur))
mean_values = np.mean(X_vel, axis=0)
mean_values = mean_values/np.max(mean_values)*100
sem_values = np.std(X_vel, axis=0) /  np.sqrt(X_vel.shape[0])
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='gray', label='X-velocities '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 4] = mean_values

Y_vel = np.abs(Chewie20161014_XYTarget[:, 1]).reshape((-1, dur))
mean_values = np.mean(Y_vel, axis=0)
mean_values = mean_values/np.max(mean_values)*100
sem_values = np.std(Y_vel, axis=0) /  np.sqrt(Y_vel.shape[0])
peak = np.round(np.max(mean_values), 1)
plt.plot(time_bins, mean_values, linewidth=2, color='black', label='Y-velocities '+str(peak)+'%')
plt.fill_between(time_bins, mean_values - sem_values, mean_values + sem_values, alpha=0.075)
mean_trace[:, 5] = mean_values
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Time from go cue (ms)')
plt.ylabel('Direction classification accuracy (%)')
# plt.title('Mean and SD across sessions: Chewie2016 PMd')
# ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.grid(False)
plt.savefig('./NER_Figs_pdf/Fig4/Chewie2016_Direction_accuracy_across_time_M1.pdf', bbox_inches='tight')
```
